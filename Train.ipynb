```python
import os
import glob


# Path to your label files directory
label_directory = '/content/Brands-1/test/labels'

# Function to update class ID in label files
def update_class_id(label_directory, old_class_id=0, new_class_id=1):
    # Get list of all label files in the directory
    label_files = glob.glob(os.path.join(label_directory, '*.txt'))

    for label_file in label_files:
        with open(label_file, 'r') as file:
            lines = file.readlines()

        with open(label_file, 'w') as file:
            for line in lines:
                parts = line.strip().split()
                if int(parts[0]) == old_class_id:
                    parts[0] = str(new_class_id)
                file.write(' '.join(parts) + '\n')

# Run the function
update_class_id(label_directory)
```
